# 9 Módulos en Python

---



Los módulos son ficheros con código Python, también se conocen como librerías.

## 9.1 Crear un módulo

Por ejemplo deseamos definir dos variables y dos funciones para que puedan ser utilizadas por otros módulos o aplicaciones de Python. Para ello las escribimos en un fichero llamado **modulo.py** 


```
variable1 = 1
variable2 = 2

def funcion1():
   print("Función 1")

def funcion2():
   print("Función 2")
```

## 9.2 Utilizar un módulo

Para utilizar variables o funciones de un módulo debemos utilizar la sentencia `import`, seguido del nombre del módulo (nombre del fichero).

```
import modulo

#Utilizar variables
print(modulo.variable1)
print(modulo.variable2)
#Utilizar funciones
modulo.funcion1()
modulo.funcion2()
```

## 9.3 Alias del módulo

Se puede definir un alias del módulo utilizando la palabra `as`

```
import modulo as mo

print(mo.variable1)
print(mo.variable2)
mo.funcion1()
mo.funcion2()
```

## 9.4 Lista de variables y funciones de un módulo

Se pueden visualizar las variables y funciones de un módulo utilizando la función `dir`.

```
import modulo

x = dir(modulo)
print(x)
```

## 9.5 Importar partes de un módulo

En lugar de importar todo el módulo, se puede importar parte del mismo utilizando la palabra `from` antes del módulo y después `import` del componente que se desea importar.

Por ejemplo si solo deseo importar la `funcion1()`:

```
from modulo import funcion1

funcion1()
```
