# 7 Clases y objetos en Python

---



En este séptimo Notebook se muestran las clases y los objetos en Python.

Python es un lenguaje orientado a objetos. En Python practicamente todo son objetos, con sus propiedades y métodos.


## 7.1 Definición de una clase

Una **clase** en Python se define con la palabra reservada **class**.

```
class MiClase:
  p1 = 1
  p2 = 2
```

## 7.2 Creación de un objeto

Una vez definida una clase, se puede crear un **objeto** que es una instancia de la clase.

Una vez instanciada la clase, se puede acceder a las **propiedades** del objeto.

```
objeto1 = MiClase()

print(objeto1.p1)
print(objeto1.p2)
```

## 7.3 Función de construcción __ init__()

La función de construcción __init__(), se ejecuta cuando de instancia la clase y su utilidad es inicializar propiedades del objeto.

```
class Persona:
  def __init__(self, nombre, apellido, edad):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = edad

persona1 = Persona("Antonio", "Pérez", 31)

print(persona1.nombre)
print(persona1.apellido)
print(persona1.edad) 
```

## 7.4 Métodos de la clase

Una clase además de definir un conjunto de **propiedades**, define un conjunto de **métodos**, los métodos son funciones que generalmente leen o modifican las propiedades de un objeto.

```
class Persona:
  def __init__(self, nombre, apellido, edad):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = edad

  def fpropiedades(self):
    print("Las propiedades son:", self.nombre, self.apellido, self.edad)

persona1 = Persona("Antonio", "Pérez", 31)
persona1.fpropiedades() 
```

## 7.5 El parámetro self

El parámetro **self** se utiliza dentro de las funciones constructoras y los métodos de la clase e indica que se accede a una propiedad del objeto.

El parámetro self puede tener cualquier otro nombre válido, por ejemplo *this*.

```
class Persona:
  def __init__(this, nombre, apellido, edad):
    this.nombre = nombre
    this.apellido = apellido
    this.edad = edad

  def fpropiedades(this):
    print("Las propiedades son:", this.nombre, this.apellido, this.edad)

persona1 = Persona("Pedro", "López", 35)
persona1.fpropiedades() 
```

## 7.6 Modificación de propiedades de un objeto

Se pueden modificar las propiedades de un objeto, asignando un valor a la propiedad.

```
print("Antes:", persona1.edad)
persona1.edad = 32
print("Despues:", persona1.edad)
```

## 7.7 Eliminar propiedades de un objeto

Se pueden eliminar propiedades de un objeto, utilizando la palabra **del**.

Al intentar acceder a una propiedad eliminada Python mostrará un error.

```
print("Antes:", persona1.edad)
del persona1.edad
print("Despues:", persona1.edad)
```

## 7.8 Eliminar objetos

Se pueden eliminar objetos, utilizando la palabra **del**.

Al intentar acceder a un objeto eliminado Python mostrará un error.

```
print("Antes:", persona1.nombre)
del persona1
print("Despues:", persona1.nombre)
```
