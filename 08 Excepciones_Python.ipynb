## 8.1 Definición de una excepción

Cuando se produce un error. Python detiene la ejecución y lanza un error.

```
print(x)
```

Para evitar que el programa se detenga por un error se define un bloque `try` en el que estarán las instrucciones que potencialmente pueden generar errores y se define un bloque `except` en el que se mostrará normalmente un mensaje de error.

```
try:
  print(x)
except:
  print("La variable x no está definida") 
```

## 8.2 Definición de varias excepciones

Si se pueden producir varios errores, se controlan en un bloque `except` cada uno.

```
try:
  print(x)
except NameError:
  print("La variable x no está definida")
except:
  print("Excepción genérica") 
```

## 8.3 Bloque else

Se puede añadir un bloque `else` que se ejecuta si no hay excepciones.

```
y = 1
try:
  print(y)
except:
  print("Bloque except")
else:
  print("Bloque else") 
```

```
try:
  print(x)
except:
  print("Bloque except")
else:
  print("Bloque else") 
```

## 8.4 Bloque finally

Se puede añadir un bloque `finally` que se ejecuta siempre, haya o no excepciones.

```
try:
  print(x)
except:
  print("La variable x no está definida")
finally:
  print("Bloque finally") 
```

El bloque `finally` es útil en el uso de ficheros, para cerrar en caso de error.

```
try:
  f = open("fichero.txt")
  f.write("xxx")
except:
  print("Error al escribir en el fichero")
finally:
  f.close() 
```

## 8.5 Lanzar una excepción

Para lanzar una excepción se utiliza la palabra reservada `raise` y a continuación el tipo de excepción que se desea lanzar

```
a = -1

if a < 0:
  raise Exception("Número menor que 0") 
```

Se pueden modificar las propiedades de un objeto, asignando un valor a la propiedad.

```
a = "ABC"

if not type(a) is int:
  raise TypeError("Solo se permiten números enteros") 
```
