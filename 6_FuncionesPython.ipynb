# 6 Funciones en Python

En este sexto Notebook se muestran las funciones en Python.

Las funciones son módulos de código que realizan una determinada funcionalidad.

## 6.1 Definición de funciones

Una **función** en Python se define con la palabra reservada **def**.

```
def mifuncion():
  print("Mi primera función")
```

## 6.2 Llamada a una función

La llamada a una función se realiza escribiendo el nombre de la función seguido de paréntesis **()** y dentro de los paréntesis los argumentos si los hubiera.

```
mifuncion()
```

## 6.3 Parámetros y argumentos de una función

Una función puede recibir datos, que se conocen como parámetros en la definición y argumentos en la llamada. Se indican entre los parentesis de la función, separados por comas y pueden ser de cualquier tipo básico o colección.

En la llamada se deben pasar el mismo número de argumentos que parámetros se han definido en la función.

```
def mifuncion(nombre, edad, puntos):
  print("Primer parámetro: ", "nombre:", nombre)
  print("Segundo parámetro:", "edad:  ", edad)
  print("Tercer parámetro: ", "puntos:", puntos)

mifuncion("Antonio", 25, 12.5)
```

Se pueden pasar en la llamada los argumentos en un orden distinto al de la definición de parámetros indicando el nombre del parámetro seguido del argumento asociado, es decir en modo clave-valor. 

```
def mifuncion(nombre, edad, puntos):
  print("Primer parámetro: ", "nombre:", nombre)
  print("Segundo parámetro:", "edad:  ", edad)
  print("Tercer parámetro: ", "puntos:", puntos)

mifuncion(edad=25, nombre="Antonio", puntos=12.5)
```

Se pueden pasar colecciones de datos como argumento de una función.

```
def imprimelista(lista):
  for x in lista:
    print(x)

listacompra = ["manzana", "plátano", "fresa"]

imprimelista(listacompra)
```

## 6.4 Número indefinido de argumentos

En el caso de no saber cuántos argumentos se van a pasar a la función, se coloca un asterisco * como prefijo del nombre del parámetro en la definición de la función.

De este modo la función recibirá una tupla como argumento.

```
def damenombres(*nombres):
  print("La tupla recibida es:", nombres)

damenombres("Antonio", "José", "Pedro") 
```

En el caso de no saber cuántos argumentos en modo clave-valor se van a pasar a la función, se coloca un doble asterisco ** como prefijo del nombre del parámetro en la definición de la función.

De este modo la función recibirá un diccionario como argumento.

```
def dameregistro(**registro):
  print("El diccionario recibido es:", registro)

dameregistro(nombre = "Antonio", edad = 21, puntos=21.5)
```

## 6.5 Parámetros por defecto

Se puede establecer un valor determinado a un parámetro si se asigna un valor al mismo en la definición de la función.

Ese valor por defecto del argumento será el valor del mismo si no se indica ese argumento en la llamada.

```
def totalproductos(numero = 0):
  print("Productos recibidos:", numero)

totalproductos()
totalproductos(1)
totalproductos(2)
```

## 6.6 Retorno de una función

Para que una función retorne un valor, se utiliza la palabra reservada **return** antes del valor que va a retornar.

```
def opsuma(a, b):
  return a + b
 
print(opsuma(1, 2))
print(opsuma(1.5, 4.1))
print(opsuma("abc", "def"))
```

## 6.7 Funciones lambda

Una función lambda es una pequeña función anónima. Puede recibir cualquier número de argumentos y solo puede tener una expresión.

```
f = lambda a : a * 2
print(f(2)) 
```

```
f = lambda a, b : a + b
print(f(1, 2)) 
```

```
f = lambda a, b, c : a + b + c
print(f(1, 2, 3)) 
```

Las funciones lambda son útiles dentro de otras funciones.

```
def multiplicador(n):
  return lambda a : a * n

doble = multiplicador(2)
triple = multiplicador(3)

print(doble(5))
print(triple(5))
```
