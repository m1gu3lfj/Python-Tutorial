# 3 Operadores en Python

---



En este tercer Notebook se muestran los operadores de Python.

Python tiene una serie de operadores para realizar operaciones entre variables y valores:

Los operadores se clasifican en los siguientes grupos:

* Aritméticos
* Comparación
* Lógicos
* Identidad
* Pertenencia
* Bits
* Asignación


## 3.1 Operadores aritméticos

```
a = 5
b = 2

print("Suma:     ", a + b)
print("Resta:    ", a - b)
print("Producto: ", a * b)
print("División: ", a / b)
print("Div. Ent: ", a // b)
print("Resto Div:", a % b)
print("Potencia: ", a ** b)
```

## 3.2 Operadores de comparación

```
a = 5
b = 2

print("a igual que b:        ", a == b)
print("a distinto que b:     ", a != b)
print("a mayor que b:        ", a > b)
print("a menor que b:        ", a < b)
print("a mayor o igual que b:", a >= b)
print("a menor o igual que b:", a <= b)
```

## 3.3 Operadores lógicos

```
a = True 
b = False 

print("a y b:", a and b)
print("a o b:", a or b)
print("no a: ", not a)
print("no b: ", not b)
```

## 3.4 Operadores de identidad

```
a = 1
b = 2
c = a

print("a es b   :", a is b)
print("a no es b:", a is not b)
print("a es c   :", a is c)
print("a no es c:", a is not c)
```

## 3.5 Operadores de pertenencia

```
numeros_impares = [1, 3, 5, 7, 9]
numeros_pares = [2, 4, 6, 8, 10]

print("1 pertenece a numeros_pares", 1 in numeros_impares)
print("1 no pertenece a numeros_pares", 1 not in numeros_impares)
print("2 pertenece a numeros_pares", 2 in numeros_pares)
print("2 no pertenece a numeros_pares", 2 not in numeros_pares)
```

## 3.6 Operadores de bits

```
a = 3 #0011
b = 9 #1001

print("a y b:                ", a & b) #0001 
print("a o b:                ", a | b) #1011 
print("a o(exclusivo) b:     ", a ^ b) #1010
print("inversión de a:       ",~ a) #1100
print("a desplaza izq. 1 bit:", a << 1) #0110
print("a desplaza der. 1 bit:", a >> 1) #0001
```

## 3.7 Operadores de asignación

```
a = 5
print("a =", a)
a += 1
print("a + 1 =", a)
a -= 1
print("a - 1 =", a)
a *= 2
print("a * 2 =", a)
a /= 2
print("a / 2 =", a)
```
